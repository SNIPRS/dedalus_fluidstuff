```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

c = 0.5
z = np.linspace(0, 4, 200)
f = np.exp(-z)
g = c*z-1
h = 1-c*z

plt.plot(z,f)
plt.plot(z,g)
plt.plot(z,h)

c1 = idx = np.argwhere(np.diff(np.sign(f - g))).flatten()
plt.plot(z[c1], f[c1], 'ro')

c2 = idx = np.argwhere(np.diff(np.sign(f - h))).flatten()
plt.plot(z[c2][1], f[c2][1], 'ro')

plt.xlabel()
plt.plot(z, z*0, 'k')
```

```python


ek = z[c2][1]
ok = z[c1][0]

print(ek, ok)

eB=1
eA = eB*(np.exp(ek)+1)

oB=1
oA = oB*(np.exp(ok)-1)
x0 = np.linspace(-2,-0.5)
x1 = np.linspace(-0.5,0.5)
x2 = np.linspace(0.5, 2)


plt.plot(x0, eA*np.exp(ek*x0),'k-')
plt.plot(x1, eB*(np.exp(-ek*x1) + np.exp(ek*x1)),'k-')
plt.plot(x2, eA*np.exp(-ek*x2),'k-')
```

```python
def E(l):
    alpha = 2
    # hbar/(2m)=1 
    
    c = 1/(alpha*l)
    
    z = np.linspace(0, 20, 2000)
    f = np.exp(-z)
    g = c*z-1 #even
    h = 1-c*z #odd
    
    
    c1 = idx = np.argwhere(np.diff(np.sign(f - g))).flatten() #even
    c2 = idx = np.argwhere(np.diff(np.sign(f - h))).flatten() #odd
    
    ok = z[c2][1] 
    ek = z[c1][0]
    
    eE = -ek**2/2
    oE = -ok**2/2
    
    return (eE, oE)

def Ee(l):
    alpha = 2
    # hbar/(2m)=1 
    
    c = 1/(alpha*l)
    
    z = np.linspace(0, 20, 2000)
    f = np.exp(-z)
    g = c*z-1 #even
    h = 1-c*z #odd
    
    
    c1 = idx = np.argwhere(np.diff(np.sign(f - g))).flatten() #even
    # c2 = idx = np.argwhere(np.diff(np.sign(f - h))).flatten() #odd
    
    # ok = z[c2][1] 
    ek = z[c1][0]
    
    eE = -ek**2/2
    # oE = -ok**2/2
    
    return eE

print(E(8))
```

```python
l = np.linspace(0.6,7,50)
el = np.linspace(0.0002,7,50)
eE = []
oE = []
for x in l:
    e, o = E(x)
    # eE.append(e)
    oE.append(o)

for x in el:
    e = Ee(x)
    eE.append(e) 
    
plt.plot(el, eE, label='even energies')
plt.plot(l, oE, label='odd energies')
plt.legend()
plt.ylabel('Energy')
plt.xlabel('l')
plt.show()
```
