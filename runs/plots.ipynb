```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

prefixes = ['02_02_', '02_1_', '02_5_', '1_02_', '1_1_', '1_5_', '5_02_', '5_1_', '5_5_']
names = ['grad', 'grad_rad', 'Tstop', 'Tgrow', 'Tsettl', 'Zgr', 'kappa_gr', 'size',
        'rho_gas_cbrt', 'Temperature', 'Press', 'rho_gas_norm', 'Temperature_norm',
        'Env', 'grad_mu', 'grad_ad', 'grad_diff', 'log10(R_0)', 'lambda_mfp', 'c_gas',
        'nu_visc', 'xi_rad', 'kappa_tot', 'Tfourth', 'rho_gas', 'Prandtl_number', 
         'tau_opt', 'Ledoux']


def plot_versions(n):
    name = names[n]
    for prefix in prefixes:
        R = np.loadtxt('common/'+prefix+'R.txt')
        var = np.loadtxt('common/'+prefix+name+'.txt')
        plt.plot(R, var, label=prefix.split('_')[0]+'Au '+prefix.split('_')[1]
                +'Earth Mass')
    plt.title(name)
    plt.legend()
```

```python
n = len(names)
plt.figure(figsize=(8, 8*n))
for i in range(n):
    plt.subplot(n,1,i+1)
    plot_versions(i)
```
